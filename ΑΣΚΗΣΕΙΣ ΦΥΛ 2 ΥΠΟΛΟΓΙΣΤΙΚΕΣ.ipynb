```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
h=0.1
N=(2-0)/h
```

```python
t= np.linspace(0,2,int(N))
t
```

```python
y=1000/(9*np.exp(-10*t)+1)
y
```

#### ΑΝΑΛΥΤΙΚΗ ΛΥΣΗ 

```python
plt.plot(t,y, label='analytical')
plt.legend()
plt.show()
```

#### ΜΕΘΟΔΟΣ EULER 

```python
yeuler= np.zeros(int(N))
yeuler[0]=100
for i in range (0,int (N-1)):
    yeuler[i+1]=yeuler[i]+h*10*yeuler[i]*(1-yeuler[i]/1000)
yeuler    
```

```python
plt.plot(t,y,label='analytical')
plt.plot(t,yeuler,label='euler')
plt.legend()
plt.show()
```

## ΑΣΚΗΣΗ 7

```python
#Εστω t ανήκει στο [0,2]
h=0.1
N=(2-0)/h
```

```python
t=np.linspace(0,2,int(N))
t
```

#### ΑΝΑΛΥΤΙΚΗ ΛΥΣΗ  

```python
y=np.exp(-t)
y
```

```python
plt.plot(t,y, label ='analytical')
plt.legend() 
plt.show()
```

#### ΜΕΘΟΔΟΣ EULER 

```python
yeuler = np.zeros(int(N))
yeuler[0]=1
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h* (-y[i])
yeuler
```

```python
plt.plot(t,y, label ='analytical')
plt.plot(t,yeuler, label ='euler')
plt.legend()
plt.show()
```

#### ΜΕΘΟΔΟΣ ΠΕΠΛΕΓΜΕΝΗ EULER

```python
h=0.1
N=(2-0)/h
```

```python
t = np.linspace(0,2,int(N))
t
```

```python
ypeuler = np.zeros(int(N))
ypeuler[0]=1
for i in range(0,int(N-1)):
    ypeuler[i+1] = ypeuler[i]/(1+h) 
ypeuler
```

```python
plt.plot(t,y, label ='analytical')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,ypeuler, label ='peplegmeni')

plt.legend()
plt.show()
```

#### ΜΕΘΟΔΟΣ ΤΡΑΠΕΖΙΟΥ 

```python
h = 0.1
N = (2-0)/h
```

```python
t = np.linspace(0,2,int(N))
t
```

```python
trapeziou = np.zeros(int(N))
trapeziou[0]=1
for i in range(0,int(N-1)):
    trapeziou[i+1] = (trapeziou[i]-(h/2)*trapeziou[i])*(1/(1+(h/2)))
ypeuler
```

```python
plt.plot(t,y, label ='analytical')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,ypeuler, label ='peplegmeni')
plt.plot(t,trapeziou, label ='trapeziou')


plt.legend()
plt.show()
```

## ΑΣΚΗΣΗ 9

```python
h=0.1
N = (2*math.pi -0)/h
t = np.linspace(0, 2*math.pi, int(N))
t
```

#### ΑΝΑΛΥΤΙΚΗ ΛΥΣΗ  

```python
y = np.cos(t)
y
```

```python
plt.plot(t, y, label = 'analytical')
plt.legend()
plt.show()
```

#### ΜΕΘΟΔΟΣ EULER 

```python
yeuler = np.zeros(int(N))
yeuler[0]=1
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h*(1/2 - (y[i]**2)/2)
yeuler
```

```python
plt.plot(t,y, label ='analytical')
plt.plot(t,yeuler, label ='euler')

plt.legend()
plt.show()
```

#### ΜΕΘΟΔΟΣ ΜΕΣΟΥ 

```python
ymesou = np.zeros(int(N))
ymesou[0]=1
for i in range(0,int(N-1)):
    ymesou[i+1] = ymesou[i] + h*(1/2)*(1-(ymesou[i]/2 +(1-ymesou[i]**2)/4)**2)
ymesou
```

```python
plt.plot(t,y, label ='analytical')
plt.plot(t,yeuler, label ='euler')
plt.plot(t,ymesou, label ='mesou')

plt.legend()
plt.show()
```

Παρατηρώ ότι αυτές οι προσεγγιστικές λύσεις είναι δύσκολο να προσεγγίζουν την ακριβή λύση του προβλήματος.

## ΑΣΚΗΣΗ 10 

```python
h=0.1
N = (1-0)/h
x = np.linspace(0,1,int(N))
x
```

#### ΑΝΑΛΥΤΙΚΗ ΛΥΣΗ  

```python
y = np.exp(x) -x -1 
y
```

#### ΜΕΘΟΔΟΣ EULER 

```python
yeuler = np.zeros(int(N))
yeuler[0]=0
for i in range(0,int(N-1)):
    yeuler[i+1] = yeuler[i] + h*(x[i]+y[i])
yeuler
```

```python
plt.plot(x,y, label ='analytical')
plt.plot(x,yeuler, label ='euler')
plt.legend()
plt.show()
```

#### TAYLOR 

```python
ytaylor = np.zeros(int(N))
ytaylor[0]=0
for i in range(0,int(N-1)):
    ytaylor[i+1] = ytaylor[i] + h* (x[i]+y[i]) + (h**2/2)*(1+x[i]+y[i])
ytaylor
```

```python
plt.plot(x,y, label ='analytical')
plt.plot(x,yeuler, label ='euler')
plt.plot(x,ytaylor, label ='taylor')

plt.legend()
plt.show()
```
